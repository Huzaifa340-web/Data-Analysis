# Daily Average Engagement and Attention
- Objective: Track how student attention/engagement varies daily.
- Tools: pandas, matplotlib.pyplot
- Tasks:
- Group by date and calculate average attention and engagement.
- Plot line charts for daily trends.

```python
level_map = {
    'Low': 1,
    'Moderate': 2,
    'High': 3
}

df['Engagement Score'] = df['Engagement Level'].map(level_map)
df['Attention Score'] = df['Attention Level'].map(level_map)

df['Date'] = pd.to_datetime(df['Date'])

daily_avg = df.groupby('Date')[['Engagement Score', 'Attention Score']].mean().reset_index()
print(daily_avg)

daily_avg[['Engagement Score', 'Attention Score']] = daily_avg[['Engagement Score', 'Attention Score']].round().astype(int)
print(daily_avg)

reverse_map = {
     1 : 'Low',
     2 : 'Moderate',
     3 : 'High' 
}

daily_avg =  daily_avg.replace(reverse_map)

print(daily_avg)
```

```python
level_map2 = {'Low': 1, 'Moderate': 2, 'High': 3}
daily_avg_numeric = daily_avg.replace(level_map2)

plt.figure(figsize=(10, 6))

plt.plot(daily_avg_numeric['Date'], daily_avg_numeric['Engagement Score'], label='Engagement', marker='o', color='g')
plt.plot(daily_avg_numeric['Date'], daily_avg_numeric['Attention Score'], label='Attention', marker='s', color='b')

plt.xlabel('Date')
plt.ylabel('Average Level (Low = 1, Moderate = 2, High = 3)')
plt.title('Daily Average Engagement and Attention')
plt.legend()
plt.grid(True)
plt.yticks([1, 2, 3], ['Low', 'Moderate', 'High'])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

# Most and Least Engaged Students
- Objective: Identify students who consistently perform well or poorly.
- Tools: pandas, numpy, matplotlib
- Tasks:
- Group by student_id and average engagement or attention.
- Sort and display top/bottom 10 students using a bar chart.


```python
level_map3 = {
    'Low': 1,
    'Moderate': 2,
    'High': 3
}

df['Engagement Score'] = df['Engagement Level'].map(level_map3)
df['Attention Score'] = df['Attention Level'].map(level_map3)

a = df.groupby('Student ID')[['Engagement Score', 'Attention Score']].mean().reset_index()
print(a)

top_engaged = df.groupby('Student ID')['Engagement Score'].mean().sort_values(ascending=False).head(10)
a_rounded1 = top_engaged.round().astype(int)
print(a_rounded1)

least_engaged = df.groupby('Student ID')['Engagement Score'].mean().sort_values().head(10)
a_rounded2 = least_engaged.round().astype(int)
print(a_rounded2)

reverse_map2 = {
     1 : 'Low',
     2 : 'Moderate',
     3 : 'High' 
}

top_engaged = a_rounded1.replace(reverse_map2)

print(top_engaged)

least_engaged = a_rounded2.replace(reverse_map2)

print(least_engaged)
```

```python
level_map4 = {'Low': 1, 'Moderate': 2, 'High': 3}
top_engaged_numeric = top_engaged.map(level_map4)

plt.figure(figsize=(10, 5))
colors = [
    'mediumseagreen',  # Fresh green
    'goldenrod',       # Warm yellow
    'tomato',          # Energetic red
    'steelblue',       # Calm blue
    'orchid',          # Soft purple
    'dodgerblue',      # Bright blue
    'darkorange',      # Vibrant orange
    'slategray',       # Neutral gray
    'teal',            # Balanced green-blue
    'royalblue'        # Bold blue
]
top_engaged_numeric.plot(kind='bar', color=colors)

plt.yticks([1, 2, 3], ['Low', 'Moderate', 'High'])

plt.title('Top 10 Engaged Students')
plt.xlabel('Student ID')
plt.ylabel('Engagement Status (High)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

least_engaged_numeric = least_engaged.map(level_map4)

plt.figure(figsize=(10, 5))
colors = ['mediumseagreen', 'goldenrod', 'tomato', 'steelblue', 'orchid'] * 2
least_engaged_numeric.plot(kind='bar', color=colors)

plt.yticks([1, 2, 3], ['Low', 'Moderate', 'High'])

plt.title('Top 10 Engaged Students')
plt.xlabel('Student ID')
plt.ylabel('Engagement Status (Low)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Behavioral Pattern Analysis
## Engagement by Learning Mode
- Objective: Analyze which learning modes lead to higher engagement.
- Tools: pandas, matplotlib
- Tasks:
- Group by learning_mode and compute average engagement.
- Plot a bar chart to compare modes.

```python
a= df.groupby('Learning Mode Prediction')['Engagement Score'].mean().sort_values(ascending=False)
print(a)


plt.figure(figsize=(8, 5))
a.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Engagement by Learning Mode')
plt.xlabel('Learning Mode')
plt.ylabel('Average Engagement')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
